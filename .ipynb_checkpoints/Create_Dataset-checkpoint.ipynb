This is the first step in Sentiment Analysis using BERT. In this notebook, we will create a dataset of reviews of applications on Google Pay Store. We will take 15 applications' review data, and data will be balance with respect to sentiments.

```python
#Import the libraries

import json
import pandas as pd
import numpy as np
from tqdm import tqdm

#Visualizaing
import matplotlib.pyplot as plt
import seaborn as sns

#Handle json data
from pygments import highlight
from pygments.lexers import JsonLexer
from pygments.formatters import TerminalFormatter

#Scrapping data
from google_play_scraper import Sort, reviews, app
```

The data is of top applications from Productivity category of Google Pay Store. Here, apps are selected based on AppAnnie rankings.

```python
#Make a list of app's ids
app_packages = [
  'com.anydo',
  'com.todoist',
  'com.ticktick.task',
  'com.habitrpg.android.habitica',
  'cc.forestapp',
  'com.oristats.habitbull',
  'com.levor.liferpgtasks',
  'com.habitnow',
  'com.microsoft.todos',
  'prox.lab.calclock',
  'com.gmail.jmartindev.timetune',
  'com.artfulagenda.app',
  'com.tasks.android',
  'com.appgenix.bizcal',
  'com.appxy.planner'
]
```

Scrape the information of each app

```python
app_infos = []

for ap in tqdm(app_packages):

  #Get the info using app method of google_play_scrapper
  info = app(ap, lang='en', country='us')
  
  #Remove the comments as of now
  del info['comments']
  
  app_infos.append(info)
```

```python
#Define a helper function to view the collected information in JSON
def print_json(json_object):
  json_str = json.dumps(json_object, indent=2, sort_keys=True, default=str)
  
  #Use pygments methods to view the source code in json in a readable way
  print(highlight(json_str, JsonLexer(), TerminalFormatter()))
```

```python
#Test the function to view JSON
print_json(app_infos[0])
```

Lets view the icons of the apps

```python
#Define a function to format the title
def format_title(title):

  #Get the separator index
  sep_index = title.find(':') if title.find(':') != -1 else title.find('-')

  if sep_index != -1:
    title = title[:sep_index]
  return title[:10]
```

```python
#View the icons of 15 apps
fig, ax = plt.subplots(3, 5, figsize = (15, 5))

for i, ax in enumerate(ax.flat):
  ai = app_infos[i]
  img = plt.imread(ai['icon'])
  ax.imshow(img)
  ax.set_title(format_title(ai['title']))
  ax.axis('off')
```

Store all the app information in a csv file

```python
app_infos_df = pd.DataFrame(app_infos)
app_infos_df.to_csv('appInfo.csv', index=False, header=True)
```

Scrape the reviews (balanced and for all the apps)

```python
app_reviews = []

for ap in tqdm(app_packages):
  for score in range(1, 6):   #Ratings score
    for sort_order in [Sort.MOST_RELEVANT, Sort.NEWEST]:
      rvs, _ = reviews(
          ap, 
          lang='en',
          country='us',
          sort=sort_order,
          count=200 if score == 3 else 100,
          filter_score_with=score
      )

      for r in rvs:
        r['sortOrder'] = 'most_relevant' if sort_order==Sort.MOST_RELEVANT else 'newest'
        r['appId'] = ap
      app_reviews.extend(rvs)
```

```python
#Print a sample review
print_json(app_reviews[0])
```

```python
#Save the reviews in csv
app_reviews_df = pd.DataFrame(app_reviews)
app_reviews_df.to_csv('appReviews.csv', index=False, header=True)
```
